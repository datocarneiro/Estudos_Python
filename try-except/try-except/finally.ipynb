## ***Cdigo correto tratando a exceo***

`````python
try:
    numero = int(input("digite um numero inteiro: "))
    print(numero)
except ValueError as e: 
    print('O valor informado no  um numero inteiro', {e})

# resultado: "O valor informado no  um numero inteiro"
`````

# ***Passando duas excees***

 possivel passar duas excees ao mesmo tempo para tartamento

para isso `(passamos um tupla)` com as excees 

`````python
try:
    numero = int(input("digite um numero inteiro: "))
    print(numero)
except (ValueError, IndexError) as e: 
    print('O valor informado no  um numero inteiro', {e})

# resultado: "O valor informado no  um numero inteiro"
`````

# ***Funo isinstance()***
A funo isinstance() em Python  usada para `verificar se um objeto  uma instncia de um determinado tipo de dados`. Seu uso  mais amplo do que apenas verificar se um objeto  uma instncia de uma classe especfica;

tambm pode ser usado para `verificar se um objeto  uma instncia de um tipo de dados bsico, como int, str, list, etc.`

Por exemplo, voc pode us-lo para garantir que uma varivel contm um determinado tipo de dado antes de realizar operaes com ela. Isso ajuda a evitar erros de tipo durante a execuo do cdigo.

Aqui est um exemplo simples:

`````python
x = 10

if isinstance(x, int):
    print("x  um inteiro.")
else:
    print("x no  um inteiro.")

# resultado: x  um inteiro.
`````
