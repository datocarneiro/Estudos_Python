# ***Formas de criar um set***

## **Com a funo set():**

`Ateno:` no garante a ordem dos elemento (no tem ndice)

`Dica:` passar tupla, lista como parametro 

````python
set1 = set()
print(set1, type(set1))
# resultado:    set() <class 'set'>

set1 = set('1,2,3,4,5')
print(set1, type(set1))
# resultado:    {'5', '3', ',', '4', '2', '1'} <class 'set'>

set2 = set('Dato')
print(set2, type(set2))
# resultado:    {'o', 't', 'a', 'D'} <class 'set'>

'''para no ficar o elemento iteravl sobre cada letra, posso passar um iteravel dentro (tupla, lista) '''
set3 = set(('carneiro',1 ,2, 3, 4))
print(set3, type(set3))
# resultado:    {1, 2, 3, 4, 'carneiro'} <class 'set'>
````

## **Com a estrutura set = {}** 

    #CUIDADO: Para criar um set vazio use a funo set(), se no vai acabar criando um dicionario, pois sua estrutura  muito semelhante

````python
set3 = {'1,2,3,4,5'}
print(set3, type(set3))
# resultado:    {'1,2,3,4,5'} <class 'set'>

set4 = {'Dato'}
print(set4, type(set4))
# resultado:    {'Dato'} <class 'set'>
````

# ***Entendendo melhor um set***

## **Um set, remove automaticamente elementos duplicados**

mas no garante a ordem dos elementos no retorno

````python
set1 = {10, 20, 20, 20, 20, 30, 40}
print(set1)
# resultado:    {40, 10, 20, 30}
````

## **set no possuem indixes**

mas podemos saber se um elemento existe da seguinte forma 

````python
set1 = {1,2,3,4,5,6}
print(5 in set1)
# resultado :   True

print(7 in set1)
# resultado :   False
````

## ***Mtodo add***

Adiciona um elemento ao set

````python
set1 = {1,2}
print(set1)
# resultado :   {1, 2}

set1.add('Dato')
print(set1)
# resultado :   {1, 2, 'Dato'}
````

## ***Mtodo update***

semelhante ao mtodo add, mas a funo  atualizar o set

````python
set1 = {1,2}
print(set1)
# resultado :   {1, 2}

set1.update(('Dato',))
print(set1)
# resultado :   {1, 2, 'Dato'}
````

## ***Mtodo Discard***
Descartando/deletando um elemento com mtodo discard

`Syntax`
```python
set.discard('elemento') # passar o elemento que ser discartado como argumento
```

````python
set1 = {1,2,'Dato', 3, 4, 5}
print(set1)
# resultado :   {1, 2, 3, 4, 5, 'Dato'}

set1.discard('Dato')
print(set1)
# resultado :   {1, 2, 3, 4, 5}
````

## ***Mtodo Clear*** 
Apagando `todos os elementos` com o mtodo clear()

````python
set1 = {1,2,'Dato', 3, 4, 5}
print(set1)
# resultado :   {1, 2, 3, 4, 5, 'Dato'}

set1.clear()
print(set1)
# resultado :   set()
````

## ***' |'  ... unio entre dois set***

    ir retornar a unio de dois set (eliminando valores duplicados)


````python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = set1 | set2
print(set3)
# resultado :   {1, 2, 3, 4}
````

## ***'&'... Interseco***

    ir retornar somente os elementos iguais entre os dois set

````python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = set1 & set2
print(set3)
# resultado :   {2, 3}
````

## ***'-' ... diferena***
    ir retornar somente os elemento diferente entre os set, e que esto no primeiro set passado (o da esquerda)

````python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = set1 - set2
print(set3)
# resultado :   {1}

set4 = set2 - set1
print(set4)
# resultado :   {4}
````

## ***'^" ... simtrica***
    ir retornar somente os elementos que so diferentes entre os set (no esto em ambos os set)

````python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set3 = set1 ^ set2
print(set3)
# resultado :   {1, 4}
````
